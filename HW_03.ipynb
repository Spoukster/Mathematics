#### Задание 1
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10). Напишите код на Python, реализующий расчет длины вектора, заданного его координатами

```python
x = [10, 10, 10]
def length(vector):
    coord_sum = 0
    coord_count = len(vector)
    for i in range(coord_count):
        coord_sum += vector[i] ** 2
    return np.sqrt(coord_sum)
vector_length = length(x)
print('Длина вектора:', vector_length)
```

#### Задание 3
Напишите код на Python, реализующий построение графиков:
1. окружности
2. эллипса
3. гиперболы

Окружность с центром в начале координат и радиусом 5.

```python
r = 5
a = 0
b = 0
x = np.linspace (-5, 5 , 101)
y1 = []
y2 = []
for x_ in x:
    y1.append(np.sqrt((r ** 2) - (x_ - a) ** 2) + b)
    y2.append(-np.sqrt((r ** 2) - (x_ - a) ** 2) + b)
plt.xlim(-15, 15)
plt.ylim(-10, 10)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

Эллипс с центром в точке (1, 1), a=5, b=3.

```python
a = 5
b = 3
x_0 = 1
y_0 = 1
x = np.linspace (-4, 6 , 101)
y1 = []
y2 = []
for x_ in x:
    y1.append(np.sqrt((1 - ((x_ - x_0) ** 2 / a ** 2)) * (b ** 2)) + y_0)
    y2.append(-np.sqrt((1 - ((x_ - x_0) ** 2 / a ** 2)) * (b ** 2)) + y_0)
plt.xlim(-15, 15)
plt.ylim(-10, 10)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(0, 10, 100)
k1 = 1
k2 = 4
a1 = 3
a2 = 0
b1 = 4
b2 = 3
y1 = k1 * np.cos(x - a1) + b1
y2 = k2 * np.cos(x - a2) + b2
plt.plot(x, y1, label='k=1')
plt.plot(x, y2, label='k=4')
plt.legend(loc='upper right')
plt.grid()
```
